```python
wsmean = 0.260646119600057
wssix = 0.18835586232221516
wset = 0.4107912742975991

wqmean = 0.6153009163035911
wqsix = 0.46801802186936015
wqet = 0.814229490804421

wTmean = 0.7270865814807954
wTsix = 0.4023136544161131
wTet = 0.8644861963973716


hsmean = 0.47497185918367946
hssix = 0.32060420222245933
hset = 0.5707482404372985

hqmean = 0.8103265609044499
hqsix = 0.6722267555565598
hqet = 0.8957087354321445

hTmean = 0.5073743461100477
hTsix = 0.26878605391432375
hTet = 0.7306080344822644
```

```python
def find_pm(mean, sixteenth, eightyfourth):
    uplim = eightyfourth-mean
    lowlim = mean-sixteenth
    
    return mean, uplim, lowlim
```

```python
warm_s = find_pm(wsmean, wssix, wset)
warm_q = find_pm(wqmean, wqsix, wqet)
warm_T = find_pm(wTmean, wTsix, wTet)
```

```python
hot_s = find_pm(hsmean, hssix, hset)
hot_q = find_pm(hqmean, hqsix, hqet)
hot_T = find_pm(hTmean, hTsix, hTet)
```
